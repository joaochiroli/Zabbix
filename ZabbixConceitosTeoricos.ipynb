# Zabbix Conceitos teóricos

Zabbix é uma plataforma utilizada para monitoramento de ativos de rede em geral. Com ela é possível saber como está a saúde do seu ambiente computacinal, monitorando recursos de rede, cpu, memória, disco entre outros, e além disso ela utilizada para alertar de forma automatizada as equipes ou pessoas responsáveis que irão tomar alguma atitude se o recurso demonstrar alguma atividade anormal. 

# Arquitetura Geral

A solução é composta por vários componentes de software sendo eles:

    * Zabbix Server: o componente central da solução, é ele o responsável por calcular e receber métricas. 
    * Banco de Dados: local onde ficam armazenados os dados. 
    * Servidor Web: solução para interface web.
    * Zabbix Proxy: intermediário de coleta de dados, entre um ativo de rede monitorado e o Zabbix Server, ele é fundamental para o monitoramento, porque através dele é possível garantir o monitoramento de ativos que usam protocolo TCP e UDP (Firewall, roteadores, entre outros), e se por acaso houver algum problema no ambiente do cliente os dados coletados não se perdem, eles ficam armazenados na base de dados do proxy, até serem enviados ao servidor. 
    * Zabbix Agent: é instalado em servidores que serão monitorados é através dele que se faz a coleta da métricas. 
    * Monitoramento sem Agent: neste monitoramento é usado normalmente o protocolo SNMP, esta estrtégia de monitoramento é normalmente abordada para monitorar equipamentos que não se pode instalar o Agent, como: Firewalls, Roteados, Switches, entre outros. 




    
  

# Arquitetura Principal









A sua arquitetura principal é constituida de 3 serviços primordiais, já citados acima, sendo eles o serviço de Banco de Dados, o serviço do Zabbix Server e o serviço da Interface Web, não necessariamente estes 3 serviços precisam estar configurados em um mesmo servidor, você pode instalar estes 3 serviços em 3 servidores diferentes para diminuir a carga de trabalho de um servidor centralizado. 

Na maioria dos casos, grandes empresas que possuem o monitoramento de uma maneira escalável possuem estes 3 serviços configurados em host diferentes. 

OBS: a instalação do Zabbix Server só pode ser feia em SO Linux.

## Arquitetura do Zabbix Proxy 



O Proxy pode ser de dois tipos, ATIVO E PASSIVO, o ATIVO: leva os dados até o Zabbix Serve, o passivo espera o Zabbix Server vir até o proxy para coletar os dados, no meu caso eu utilizo mais o Proxy ATIVO, para tirar um pouco da carga de processamento dos servidores. 

A arquitetura de um Proxy é formada normalmente por 3 componentes o serviço Zabbix Proxy responsável por levar os dados até o Zabbix Server, o Banco de Dados que armazena os dados monitorados e o Zabbix Agent responsável por coletar métricas do proxy, O ZABBIX AGENT APENAS COLETA AS MÉTRICAS NESTE CASO, QUEM ENVIA OS DADOS É O PROXY, e neste caso é usado um Zabbix Agent de forma passiva, em que o serviço Zabbix Agent apenas coleta as métricas, o Zabbix Proxy pega as métricas e as envia para o Zabbix Server, isto será abordado de uma forma melhor no próximo tópico.


OBS: a instalação do Zabbix Server só pode ser feia em SO Unix.

## Arquitetura do Zabbix Agent



O Zabbix Agent pode ser de dois tipos, o Passivo, em que o Zabbix Agent coleta métricas ddo servidor e o serviço Zabbix Proxy pega estas métricas e leva até o Zabbix Server, ou o Ativo, em que o próprio serviço coleta e encaminha as métricas até o Zabbix Server. 

Na maioria dos casos o Zabbix Server e o Zabbix Proxy, possuem também em seu servidor o Zabbix Agent instalado para fazer coletas dos dados e saber como está a saúde operacional do sistema. 



## Arquitetura de Monitoramento contendo ativos de rede monitorados sem Agent

Já os ativos de rede que não se é possível instalar o serviço do Zabbix Agent, é utiliizado protocolos de rede para se fazer o monitoramento, como SNMP, PING, entre outros. O mais comum, é utilizar o monitoramento através de SNMPv2 ou SNMPv3 para coletar os dados como CPU, memória, tráfego em interfaces de rede, entre outros. 



Você pode achar estes dados sobre os componentes destes hosts, que são monitorados através do protocolo SNMP, por meio de identificadores de recursos disponibilizados pelas próprias fabricantes, estes identificados são chamados de OIDS, e através deles você pode coletar algum dados específico de monitoramento.
