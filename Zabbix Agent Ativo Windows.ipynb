# Install Zabbix Agent Windows

**Baixar os dados necessários para instalação**




Esses dados voce consegue acessar indo ate o site principal do Zabbix ou no Repositoio do Zabbix.

https://cdn.zabbix.com/zabbix/binaries/stable/5.0/5.0.20/zabbix_agent-5.0.20-windows-amd64.zip

**Apos baixar o pacote extraia o pacote e coloque o nome que quiser**

**Abra a pasta, dentro da pasta vao ter dois arquivos uma chamada de bin e outra de conf, entre na conf e edite o zabbix_agentd.conf**

Neste arquivo será o excluido o seguinte parametro

Server=

E sera editado os seguintes parametros 

LogFile= (coloque o caminho onde voce deseja salvar o log do service)

StartAgents=0

Hostname= (este parametro deve conter exatamente o mesmo nome do host que sera criado na interface Web do Zabbix)

ServerActive= 192.168.10.10:8081 (este parametro deve conter a porta e o ip do servico do Zabbix Server)

**Comando para instalar o servico do Zabbix Agent no Windows**

* Abrir o CMD como Administrador
* Entrar na pasta bin do arquivo criado, no meu caso seria o cd c:\Zabbix\bin
* Digitar o seguinte comando: c:\Zabbix\bin>zabbix_agentd.exe -c c:\Zabbix\conf\zabbix_agentd.conf -i -m



E agora basta ir até o aplicativo de serviços e dar o start no service de monitoramento, lembre-se de colocar o serviço para reicinar caso aconteça alguma falha, voce consegue fazer isso clicando em propriedades no serviço do Zabbix no aplicativo Serviços do Windows e clicando na aba recuperação, lá vai estar todas as opções.

Depois disso basta ir até a interface Web do Zabbix e colocar o mesmo nome que está no arquivo de configuração. No meu caso é Servidor-Windows, adicione o ip, e coloque a template, lembre-se que os itens da template neste caso precisam ser do tipo Active Agent. Se for colocar de outra forma não irá coletar os dados.
